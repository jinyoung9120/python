```python
import requests
res = requests.get('https://jgkim21.tistory.com/15708660')
result = res.text
```

```python
sidx = result.find('소나기 / 황순원')
eidx = result.find('작가소개', sidx)
result2 = result[sidx:eidx]
print(result2)
```

```python
eidx = 0
while True:
    sidx = result2.find('SPAN style="FONT-SIZE: 14pt', eidx)
    if sidx == -1: break
    sidx = result2.find('칡',sidx)
    eidx = result2.find('.',sidx)
    print(result2[sidx : eidx + 1])
```

```python
eidx = 0
while True:
    sidx = result2.find('SPAN style="FONT-SIZE: 14pt', eidx)
    if sidx == -1: break
    sidx = result2.find('칡',sidx)
    eidx = result2.find('<',sidx)
    print(result2[sidx : eidx])
```

```python
import requests
res = requests.get('http://ggoreb.com/quiz/소나기.txt')
res.encoding = None     #'utf8'   or   'ms949'
text = res.text
```

```python
위치 = 0
while True:
    위치 = text.find('칡', 위치)
    마침표 = text.find('.', 위치)

    
    if 위치 == -1: break
    print(text[위치 : 마침표 + 1].replace('\n',''))
    위치 = 위치 + 1 
```
